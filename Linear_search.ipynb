```python
def linear_search(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:
            return i
    return -1
    
arr = [2, 3 , 5, 7, 8]

target = 7

print(linear_search(arr, target))
```

```python
def count_occurences(arr, target):
    count = 0

    for i in range(len(arr)):
        if arr[i] == target:
            count += 1
        
    return count

arr = [1, 2, 3, 3, 4, 5]

print(count_occurences(arr, 3))
```

```python
def find_largest(arr):
    largest = arr[0]

    for i in range(len(arr)):
        if arr[i] > largest:
            largest = arr[i]
        
    return largest

arr = [10, 5, 32, 21 ,45, 0]

print(find_largest(arr))
```
