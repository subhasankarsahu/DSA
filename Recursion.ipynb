```python
# Factorial of a number

def factorial(n):
    if n == 0:
        return 1
    
    return n * factorial(n - 1)

print(factorial(5))
```

```python
# Sum of N Numbers

def sum_n(n):
    if n == 0:
        return 0
    
    return n + sum_n(n - 1)

print(sum_n(5))
```

```python
# Fibonacci 

def fibonacci(n):
    if n <= 1:
        return n
    
    return fibonacci(n - 1) + fibonacci(n - 2)

print(fibonacci(3))
```

```python
# Reverse a String

def reverse_string(s):
    if len(s) == 0:
        return s
    
    return reverse_string(s[1:]) + s[0]

print(reverse_string("Hello I am Subha"))
```
